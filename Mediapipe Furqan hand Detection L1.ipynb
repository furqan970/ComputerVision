```python
import mediapipe as mp
import cv2 #for webcam
import numpy as np
import uuid #allows u to generate a uniform unique indentifier to prevent overlap of images
import os
```

```python
mp_drawing = mp.solutions.drawing_utils #for drawing landmarks 
mp_hands=mp.solutions.hands
```

# Draw Hands

```python
from IPython.display import Image #just importing image for reference

Image(url='https://developers.google.com/static/mediapipe/images/solutions/hand-landmarks.png')
```

```python
cap = cv2.VideoCapture(0) #accesing webcam
with mp_hands.Hands(min_detection_confidence=0.8,min_tracking_confidence=0.5)as hands:
    while cap.isOpened():
        ret,frame= cap.read()
        #bgr to rgb
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image=cv2.flip(image,1)
        #set flags
        image.flags.writeable=False
        #detections
        results=hands.process(image)
        #set flags to true
        image.flags.writeable=True
        #BGR to RGB
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        #detections
        print(results)
        #rendering results
        if results.multi_hand_landmarks:
            for num,hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image,hand,mp_hands.HAND_CONNECTIONS,
                                          mp_drawing.DrawingSpec(color=(121,22,76),thickness=2,circle_radius=2),
                                          mp_drawing.DrawingSpec(color=(250,44,250),thickness=2,circle_radius=1)
                                         );
        cv2.imshow("Hand Tracking",image)
        if cv2.waitKey(10) &0xFF==ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

# Output Images

```python
os.mkdir("Output Images")
```

```python
cap = cv2.VideoCapture(0) #accesing webcam
with mp_hands.Hands(min_detection_confidence=0.8,min_tracking_confidence=0.5)as hands:
    while cap.isOpened():
        ret,frame= cap.read()
        #bgr to rgb
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image=cv2.flip(image,1)
        #set flags
        image.flags.writeable=False
        #detections
        results=hands.process(image)
        #set flags to true
        image.flags.writeable=True
        #BGR to RGB
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        #detections
        print(results)
        #rendering results
        if results.multi_hand_landmarks:
            for num,hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image,hand,mp_hands.HAND_CONNECTIONS,
                                          mp_drawing.DrawingSpec(color=(121,22,76),thickness=2,circle_radius=4),
                                          mp_drawing.DrawingSpec(color=(250,44,250),thickness=2,circle_radius=2)
                                         );
        cv2.imwrite(os.path.join('Output Images','{}.jpg'.format(uuid.uuid1())),image)        
        cv2.imshow("Hand Tracking",image)
        if cv2.waitKey(10) &0xFF==ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
